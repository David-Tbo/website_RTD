```python
import pptx
from pptx.util import Inches # for pictures
from pptx.enum.shapes import MSO_SHAPE # for shapes
from pptx.enum.dml import MSO_THEME_COLOR # for shapes color
from pptx.chart.data import CategoryChartData
from pptx.enum.chart import XL_CHART_TYPE
from pptx.enum.chart import XL_TICK_MARK
from pptx.util import Pt
```

## Create a new presentation

```python
# Create a new presentation
prs = pptx.Presentation()
```

## Add initial slide

```python
# Create first slide
# [0] = Title slide (title & subtitle)
# [1] = Title and content
# [2] = Section Header
# ...
```

```python
# Create the title slide:
slide1_register = prs.slide_layouts[0]

slide1 = prs.slides.add_slide(slide1_register)
```

```python
# The layout 0 that we have chosen
# allows to have a title and a subtitle
```

## Add title and subtitle

```python
# Add text (a title + subtitle) to the slide
# title

title1 = slide1.shapes.title
title1.text = "Title of the slide N°1"

# Same command but all in one
title1 = slide1.shapes.title.text = "Title of the slide N°1"

# Add a subtitle

subtitle = slide1.placeholders[1]
subtitle.text = "Subtitle of the slide N°1"

# same command in a single line
# slide1.placeholders[1].text = "Subtitle of slide N°1"
```

## Add more slides
### With bullet points

```python
# Create the bullet point slide:
slide2_register = prs.slide_layouts[1]

# Add it to the presentation:
slide2 = prs.slides.add_slide(slide2_register)

# Add the title to this slide
title2 = slide2.shapes.title
title2.text = "Title of the slide N°2"

# Create a bullet point box
bullet_point_box = slide2.shapes

# level 1: Define the first bullet point + text of level 1
bullet_points_lvl1 = bullet_point_box.placeholders[1]
bullet_points_lvl1.text = "First bullet point text"
#bullet_points_lvl1.level = 0

# level 2: Define the first bullet point + text of level 2
bullet_points_lvl2 = bullet_points_lvl1.text_frame.add_paragraph()
bullet_points_lvl2.text = "Text of the bullet point of level 2"
bullet_points_lvl2.level = 1

# level 3: Define the first bullet point + text of level 2
bullet_points_lvl3 = bullet_points_lvl1.text_frame.add_paragraph()
bullet_points_lvl3.text = "Text of the bullet point of level 3"
bullet_points_lvl3.level = 2
```

## Slide with Auto-shapes

```python
# Create the bullet point & picture slide:
slide4_register = prs.slide_layouts[5]
slide4 = prs.slides.add_slide(slide4_register)

# Add the title of this slide
title4 = slide4.shapes.title
title4.text = "Title of the slide N°4"

# Add the first shape (shape n°1)

# Positioning of the shape n°1
left1 = top1 = width1 = height1 = Inches(2)

# kind of the shape n°1: ROUND_RECTANGLE
add_shape1 = slide4.shapes.add_shape(MSO_SHAPE.ROUND_1_RECTANGLE, left1, top1, width1, height1)

# Add the second shape (shape n°2)

# Positioning of the shape n°1
left2 = Inches(6)
top2 = Inches(2)
width2 = height2 = Inches(2)

# kind of the shape n°2: ARROW
arrow1 = slide4.shapes.add_shape(MSO_SHAPE.DOWN_ARROW, left2, top2, width2, height2)

# change the main colour of the arrow
fill_arrow1 = arrow1.fill
fill_arrow1.solid() # solid colour
fill_arrow1.fore_color.theme_color = MSO_THEME_COLOR.ACCENT_2 # change the color

# rotate the arrow to 90 degrees
arrow1.rotation = 90
```

## Slide with charts

```python
# Create the bullet point & picture slide:
slide5_register = prs.slide_layouts[5]
slide5 = prs.slides.add_slide(slide5_register)

# Add the title of this slide
title5 = slide5.shapes.title
title5.text = "Title of the slide N°5"

# Build the graph
graph_info = CategoryChartData()
graph_info.categories = ["A", "B", "C"]
graph_info.add_series("Series 1", (15, 11, 18))

# Add the graph to the slide with positions
left_graph = top_graph = Inches(2)
width_graph = Inches(6)
height_graph = Inches(4)
slide5.shapes.add_chart(XL_CHART_TYPE.COLUMN_CLUSTERED, left_graph, top_graph, width_graph, height_graph, graph_info)
# we turn the above line into a variable
# # graph frame
# graph1_frame = slide5.shapes.add_chart(XL_CHART_TYPE.COLUMN_CLUSTERED, left_graph, top_graph, width_graph, height_graph, graph_info)
# graph1 = graph1_frame.chart

# # Edit graph
# category_axis = graph1.category_axis
# category_axis .has_major_gridlines = True
# category_axis.minor_tick_mark = XL_TICK_MARK.OUTSIDE
# category_axis.tick_labels.font.italic = True
# category_axis.tick_labels.font.size = Pt(24)
```

## Slide with Table

```python
# Create the bullet point & picture slide:
slide6_register = prs.slide_layouts[5]
slide6 = prs.slides.add_slide(slide6_register)

# Add the title to this slide
title6 = slide6.shapes.title
title6.text = "Title of the slide N°6"

# Add a table the graph with positioning
left_table = Inches(1)
top_table = Inches(2)
width_table = Inches(8)
height_table = Inches(3)

slide6.shapes.add_table(3, 4, left_table, top_table, width_table, height_table)
# we turn the above line into a variable
table1_frame = slide6.shapes.add_table(3, 4, left_table, top_table, width_table, height_table)

# Feed the table (cell by cell)
table1 = table1_frame.table

cell = table1.cell(0,0)
cell.text = "Insert title here"

cell = table1.cell(1,0)
cell.text = "Insert value here"

cell = table1.cell(0,1)
cell.text = "Insert another text"
```
