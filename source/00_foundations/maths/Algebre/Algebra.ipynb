## Espaces vectoriels

### Les vecteurs:

Ils peuvent représenter une force, un déplacement...  
Ils ont une norme, une direction et un sens.  

Vulguarisation:  

Un __espace vectoriel__ (ev) est un ensemble mathématiques dont les objets se comportent comme des vecteurs:
- on peut les ajouter,
- on peut changer le sens d'un objet et obtenir son opposé,
- on peut multiplier l'objet par un nombre (scalaire)

### Exemples d'espaces vectoriels:  

- $\mathbb{R}$ les points de la droite réelle
- $\mathbb{R}^n$ les n-uplets
- $\mathbb{R}^\mathbb{N}$ les suites réelles
- $M_{n,p}(K)$ les matrices $\iff \mathbb{R}^{np}$
- $K[X]$ les polynômes ($P(X) = X^2 + X + 1$ )
- $K(X)$ les fractions rationnelles ($u = \frac{X^2 +1}{X + 1}$)
- $F(X,K)$ les fonctions de X dans K
- $L(X,E)$ les applications d'un espace qcq dans un espace vectoriel

NB1: Pour $L(X,E)$, c'est l'arrivée qui compte et qui garanti que l'ensemble de ces applications est un espace vectoriel.  
NB2: Cas particulier de $L(X,E)$ quand $X = N$, les suites d'éléments de E constituent un espace vectoriel.

### __Définition d'un K-espace vectoriel__

Pour qu'un ensemble $V$ soit un $K$-espace vectoriel, il doit satisfaire les conditions suivantes:

1. Addition de vecteurs:

- 1.1 __Fermé sous l'addition__: Pour tous vecteurs $u, v \in V, u + v \in V$
- 1.2 __Associativité__: Pour tous vecteurs $u, v, w \in V, (u+v)+w = u + (v+w)$
- 1.3 __Commutativité__: Pour tous vecteurs $u,v \in V, u+v = v+u$
- 1.4 __Elément neutre__: il existe un vecteur $0 \in V$ tel que pour tout vecteur $u \in V, u+0=u$
- 1.5 __Inverse additif__: Pour tout vecteur $u \in V,$ il existe un vecteur $-u \in V $ tel que $u + (-u) = 0$

2. Multiplication scalaire:

- 2.1 __Fermé sous la multiplication scalaire__: Pour tout vecteur $u \in V$ et tout scalaire $a \in K$, $au \in V$ 
- 2.2 __Associativité__: Pour tous scalaires $a, b \in K$ et tout vecteur $u \in V, a(bu)=(ab)u$.
- 2.3 __Element neutre__: Pour tout vecteur $u \in V, 1u=u,$ où $1$ est l'élément neutre de la multiplication dans $K$.
- 2.4 __Distributivité par rapport à l'addition de vecteurs__: Pour tout scalaire $a \in K$ et tous vecteurs $u, v \in V, a(u+v)=au+av$
- 2.5 __Distributivité par rapport à l'addition de scalaires__: Pour tous scalaires $a, b \in K$ et tout vecteur $u \in V, (a+b)u=au+bu$

Note: Un $K$-espace vectoriel est une définition plus précise d'un espace vectoriel car le corps des scalaires $K$ est clairement défini.

__Comment montrer qu'un ensemble F est un un espace vectoriel__ ?  

Dans la pratique on ne montrera pas toutes les conditions de 1.1 à 2.5 sauf si l'ensemble nous est complètement inconnu.  

Il suffira de montrer que F:    

1. __Est un sous-ensemble d'un espace vectoriel E connu__
2. __Fermé sous l'addition__ 
3. __Fermé sous la multiplication__
4. __Contient l'élément neutre__

Démonstration:  

$F$ étant un sous-espace de $E$ il n'est pas nécessaire de démontrer les propriétés:  
- d'associativité (1.2 et 2.2), 
- de commutativité (1.3), 
- de l'élément neutre (2.3) et 
- de distributivité (2.4 et 2.5)  

Ces propriétés étant vraies pour tous les vecteurs de l'espace vectoriel E, elles sont en particulier vraies pour tous les vecteurs du sous-espace vectoriel F.  

Quant à la propriété de l'inverse additif (1.5), on sait que pour tout veceur $u \in F$, l'inverse additif dans $E$ est $-u$.  
Si F est fermé sous la multiplication scalaire, $-u = (-1).u \in F$  

conclusion: Si F est un sous-ensemble d'un espace vectoriel, il suffit de montrer les 3 propriétés pour conclure que F est un sous-espace vectoriel. CQFD.


__Définition d'un corps $K$:__

Un corps $K$ est un ensemble muni de deux opérations binaires, l'addition et la multiplication, qui satisfont les propriétés suivantes :

__Associativité__  
- de l'addition: $(a+b)+c = a+(b+c)$
- de la multiplication: $(a . b) . c = a . (b . c)$

__Commutativité__  
- de l'addition: $a+b = b+a$
- de la multiplication: $a . b = b . a$

__Distributivité__:  
- $a⋅(b+c)=(a⋅b)+(a⋅c)$

__Existence d'un élément neutre__:  
- pour la multiplication: $1 . a = a$
- pour l'addition: $0 + a = a$ 

__Existence d'un inverse__:  
- pour la multiplication: pour $a \neq 0$,  il existe $a^{-1}$ tel que $a . a^{-1} = 1$
- pour l'addition: pour chaque a, il existe -a tel que: $a + (-a) = 0 $ 


<u>__Propriété__:</u>  

Soit E un K-espace vectoriel et F une partie de E:  

F est un sous-espace vectoriel $\iff$:

1. F est non vide
2. F est stable par combinaison linéaire ($\text{Pour tous } a, b \in K, \text{pour tous } u, v \in F, a.u + b.v \in F$)

Démonstration:

1. Si F est non vide, alors il existe au moins un vecteur $v \in F$. En particulier, $0.v=0 \in F$, fonc $F$ contient le vecteur nul.

2. Si F est stable par combinaison linéaire:
- $\forall u,v \in F, u+v = 1.u+1.v \in F$, donc $F$ est stable par addition. 
- $\forall u \in F, -u = (-1).u \in F$, donc $F$ continent les inverses additifs.
- $\forall a \in K, u \in F, a.u \in F$, donc $F$ est stable par multiplication scalaire.  

__Exercices__

Les parties suivantes de $\mathbb{R}^2$ sont-elles des sous-espaces vectoriels ?

1. $F_1 = \{(x,y) \in \mathbb{R}^2: 2x + 3y=0\}$
2. $F_2 = \{(x,y) \in \mathbb{R}^2: x+y+1=0\}$
3. $F_3 = \{(x,y) \in \mathbb{R}^2: x^2-3y=0\}$
4. $F_4 = \{(x,y) \in \mathbb{R}^2: x^2-y^2=0\}$

Réponses:

$F_1$ est un sous-espace vectoriel de $\mathbb{R}^2$  
$0_{\mathbb{R}^2} \in F_1$ en effet $2 \times 0 + 3 \times 0 = 0$  
Soient $a, b \in \mathbb{R} \text{ et } v_1=(x_1,y_1), v_2=(x_2, y_2) \in F_1:$  
$a.v_1 + b.v_2 = (a.x_1, a.y_1)+(b.x_2, b.y_2) = (a.x_1+b.x_2, a.y_1+b.y_2) = (X, Y)$  
Et $2X + 3Y = 2(a.x_1+b.x_2) + 3(a.y_1+b.y_2) = a(2x_1+3y_1) + b(2x_2 + 3y_2) = a.0 + b.0 = 0$  

$F_2$ n'est pas un sous-espace vectoriel de $\mathbb{R}^2$ car $0_{\mathbb{R}^2} \notin F_2$.

$0_{\mathbb{R}^2} \notin F_3$ mais $F_3$ n'est pas un sous-espace vectoriel de $\mathbb{R}^2$ car $v=(1, \frac{1}{3}) \in F_3$ mais $v + v \notin F_3$.

$0_{\mathbb{R}^2} \notin F_4$ mais $F_4$ n'est pas un sous-espace vectoriel de $\mathbb{R}^2$ car $u=(-2, -2) \in F_4, v=(3, -3) \in F_4$ mais $u + v \notin F_4$.


Note: $F_4 = \{(x,y) \in \mathbb{R}^2: (x-y)(x+y)=0\}$  
$\{(x,y) \in \mathbb{R}^2: x-y=0\}$ est un sous-espace vectoriel de $\mathbb{R}^2$ et  $\{(x,y) \in \mathbb{R}^2: x+y=0\}$ est un sous-espace vectoriel de $\mathbb{R}^2$ mais l'union de ces deux sous-espaces vectoriels n'est pas un espace vectoriel.  
L'union de deux sous-espaces vectoriels forme rarement un sous-espace vectoriel.

<u>__Propriété__:</u>  

__L'intersection de 2 sous-espaces vectoriels est un espace vectoriel de E.__

Démonstration:

Soient $F$ et $G$ deux sous-espaces vectoriels de l'espace vectoriel $E$.  
- $0_E \in F, 0_E \in G \rightarrow 0_E \in F \cap G$  
- $\text{Soient } a,b \in K \text{ et } u \in F \cap G, v \in F \cap G \text{ est-ce que:  }$ 
$au + bv \in F \cap G \text{ ? } au + bv \in F, au + bv \in G, \rightarrow au + bv \in F \cap G.$

#### Les suites

__Exercice sur les parties de l'espace vectoriel des suites__ $K^\mathbb{N}$:  

Les parties suivantes sont-elles des sous-espaces vectoriels de $K^\mathbb{N} \text{ ? }$

$F_1 = \{ u = (u_n)_{n \in \mathbb{N}} \text{ telle que: } \forall n, u_{n+2} = u_{n+1} + u_n \} \subset K^\mathbb{N}$

$F_2 = \{ u = (u_n)_{n \in \mathbb{N}} \text{ telle que: } \forall n, u_{n+2} = n^3u_{n+1} + \exp(-n).u_n \} \subset K^\mathbb{N}$

Démonstration:  

Uniquement pour $F_1$, celle pour $F_2$ est laissée en exercice.

$(0,0,0,\ldots) \in F_1 \text{ en effet } 0_{K^\N} + 0_{K^\N} = 0_{K^\mathbb{N}}$  

Soient $a,b \in K$ et $u,v \in F_1$. Est-ce que $au + bv \in F_1 \text{ ? }$  
- $(au + bv)_{n+2} = au_{n+2} + bv_{n+2} = a(u_{n+1} + u_n) + b(v_{n+1} + v_n) = (au + bv)_{n+1} + (au + bv)_{n}$. Oui, $au + bv \in F_1$  

$F_1$ est bien un sous-espace vectoriel de $K^\mathbb{N}$


Exemples de sous-ensembles de suites __qui sont des espaces-vectoriels__:   

- $\{ \text{ suites constantes } \}$
- $\{ \text{ suites bornées } \}$
- $\{ \text{ suites convergentes } \}$

Exemples de sous-ensembles de suites __qui ne sont pas des sous-espaces vectoriels (s.e.v.)__:

- $\{ \text{ suites croissantes } \}$ ne sont pas des s.e.v. car en multipliant par $-1$ une suite croissante devient décroissante.
- $\{ \text{ suites monotones } \}$ n'est pas un s.e.v car par exemple: $u_n = n^2 - 4n$ est une somme de suites croissante $u:n \to n^2$ et décroissante $v:n \to -4n$  
    Hors, $u = (0,-3,-4,-3,0,5,...)$ n'est pas monotone.

#### E l'ensemble des fonctions de $\mathbb{R} \to \mathbb{R}$

- $F_1 = \{f \in E : f(0) = 0 \}$ est un sous-espace vectoriel.  
- $F_2 = \{f \in E : f(0) = 1 \}$ n'est pas un sous-espace vectoriel car ne contient pas la fonction nulle.  
- $F_3 = \{f \in E : f(0) = 0 \text{ ou } f(1) = 0\}$ n'est pas un sous-espace vectoriel car $x \to x-1 \in F_3 \text{ et } x \to x \in F_3 \text{ mais } x \to x -1+x \notin F_3$  
- $F_4 = \{f \in E : f(0) = 0 \text{ et } f(1) = 0\}$ est un sous-espace vectoriel car intersection de deux sous-espaces vectoriels.  
- $F_5 = \{f \in E : \forall t \in \mathbb{R}, \int f(t) \, dt = 0 \}$ est un sous-espace vectoriel car l'intégrale nulle vaut $0$ et les intégrales sont linéaires.  
- $F_6 = \{f \in E : f(x) = f(-x) \}$. L'ensemble des fonctions paires est un sous-espace vectoriel car la fonction nulle est paire, et 
    $ (f+g)(x) = f(x) + g(x) = f(-x) + g(-x) = (f + g)(-x)$
    $ si $a \in \mathbb{R}, af(x) = af(-x)$



Note:  

Pour ces démonstrations on a utilisé:  

- $0_E \in F$
- $\forall f \in F, g \in F, f + g \in F$
- $\forall a \in K, f \in F, af \in F$  

Plutôt que:  
- $0_E \in F$
- $\forall a,b \in K, \forall f,g \in F, af + bg \in F$  

Les deux sont équivalentes.

# Familles de vecteurs

Soit $E$ un $K$-espace vectoriel  
Soit $I$ un ensemble quelconque d'indices.

Une famille de vecteurs de $E$ indexée par $I$ est une application de $I$ dans $E$.  

$f:I \rightarrow E, \text{ posons } u_i = f(i), \text{ on note la famille } (u_i)_{e \in I}$

Exemple:  

Si $I = \{1,2 \ldots , n \}, (u_i)_{i \in I} \in E^n$ est un n-uplets de vecteurs.

## Combinaison linéaire

Soit $(u_i)_{i \in I}$ une famille de vecteurs de $E$.  

On dit qu'un vecteur **$x$** de $E$ est **combinaison linéaire de la famille** $(u_i)_{i \in I}$  

si et seulement si:  

$\exists J \subset I \text{ finie }, \exists (\alpha_i)_{i \in J}, \text{ une famille d'éléments de } K,  \text{ telle que } x = \sum_{i \in J} \alpha_i u_i$

Soit $A \in E$ une partie non vide quelconque,  

$Vect(A)$ est l'ensemble des combinaisons linéaires de familles d'éléments de $A$   

$(Vect(\emptyset)=\{0_E\})$

Nota Bene: en algèbre linéaire on travaille avec des sommes finies de combinaisons linéaires. Il n'y a pas d'histoire de convergence.

Exemples:  

1. Soit $x \in E, Vect(x) = Kx = \{ \alpha x , \alpha \in K \}$  
On a ici pris un seul vecteur $x, Vect(x)$ est l'ensemble de tous les multiples de $x$ par $\alpha$ avec $\alpha$ parcourant l'ensemble du corps commutatif $K$. On le visualise comme une droite.

2. $K^3 = Vect((1,0,0),(0,1,0),(0,0,1))$  
C'est l'ensemble des triplets de $K$. Si $(a,b,c) \in K^3 \text{ alors } (a,b,c) = a(1,0,0) + b(0,1,0) + c(0,0,1)$  
N'importe quel élément de $K^3$ est une combinaison linéaire des trois vecteurs $(1,0,0), (0,1,0), (0,0,1)$

3. $K[X] = Vect((X^n)_{n \in \mathbb{N}})$  
Les polynômes.  
Un polynôme est de la forme $\sum_{n \in \mathbb{N}} a_i X^i$ avec $n \in \mathbb{N}$ pour couvrir tous les polynômes qui sont plus nombreux qu'un nombre fini.

4. $E = M_2(K)$  
Les matrices 2x2 dans $K,  F = Vect(\begin{pmatrix}
   1 & 0 \\
   0 & 1 
\end{pmatrix}, \begin{pmatrix}
   0 & 1 \\
   -1 & 0 
\end{pmatrix})$

Pour montrer que $F$ est un espace vectoriel, on va cette fois-ci utiliser le théorême suivant:

### Espace vectoriel engendré

$\forall A \subset E, Vect(A)$ est un sous-espace vectoriel de $E$, et c'est le plus petit sous-espace vectoriel de $E$ contenant $A$ dans le sens suivant:  

Si $F$ est un sous-espace vectoriel de $E$ contenant $A$, alors $Vect(A) \subset F$

Preuve:  

1. Soit $a \in A, 0_E = 0 \times a \in Vect(A)$  (j'ai bien écrit $0$ comme une combinaison linéaire)
2. Une combinaison linéaire de combinaisons linéaires reste une combinaison linéaire  
(1 & 2): $Vect(A)$ est un sous-espace vectoriel de $E$  

3. Soit $F$ un sev de $E$ contenant $A$.  
Soit $x \in Vect(A), $ on peut écrire alors: $x = \sum \alpha_i a_i$  
$\forall i \in J, a_i \in F$ car $A \subset F$ donc $\alpha_i a_i \in F$ car $F$ est stable par combinaison linéaire en tant que sev de $E$.  
Donc $ x \in F$. On a montré que $\forall x \in Vect(A), x \in F$ i.e. $Vect(A) \subset F$.  

CQFD



Note:  
On  a pris $a \in A$ ie $A$ non vide. Mais si $A = \emptyset$ alors $Vect(A)=Vect(\emptyset)=\{0_E\}$ est bien un sev et c'est le plus petit contenant le $\emptyset$

## Famille génératrice

Soit $(u_i)_{i \in I}$ une famille de vecteurs de $E$.  

On dit que la famille $(u_i)_{i \in I}$ **est génératrice**  

si et seulement si:  

$Vect((u_i)_{i \in I} ) = E$

On peut faire tout l'espace à partir d'une famille génératrice.

Exemple 1:  

$\mathbb{R}^2 = Vect((1,0),(0,1))$

Tous les éléments de $\mathbb{R}^2$ peuvent être générés par $(1,0)$ et $(0,1)$.  
Ce sont des $\alpha(1,0) + \beta(0,1), \alpha, \beta \in \mathbb{R}$

Exemple 2:  

$\mathbb{R}^2 = Vect((3,2),(-1,1))$

Pour que deux vecteurs soit une famille génératrice de $\mathbb{R}^2$ il suffit que les deux vecteurs ne soient pas parallèles.

Exemple 3:  

Soient $u = (2,-1,3), v = (-1,4,-3), w = (1,0,1) \in \mathbb{R}^3$  

Pour montrer que $Vect((u,v,w))$ est une famille génératrice, on va montrer que $(1,0,0),(0,1,0),(0,0,1)$ sont des éléments de $Vect(\{{(u,v,w)}\})$  
Ainsi $Vect((1,0,0),(0,1,0),(0,0,1)) \subset Vect(u,v,w)$  

$4(u - 2w) + v + w = (0,0,2)$ donc $(0,0,1) \in Vect(u,v,w)$  
$w - (0,0,1) = (1,0,0)$ donc $(1,0,0) \in Vect(u,v,w)$  
$2(1,0,0) + 3(0,0,1) - u = (0,1,0) \in Vect(u,v,w)$  

Cette méthode est plus rapide que de faire $\alpha u + \beta v + \gamma w$ représente tous les éléments de $\mathbb{R}^3$

## Famille libre

Soit $(u_i)_{i \in I}$ une famille de vecteurs de $E$.  

On dit que la famille $(u_i)_{i \in I}$ **est liée**  

si et seulement si:  

$\exists j \in I, u_j \in Vect((u_i)_{i \in I, i \ne j} ) = E$

Une famille qui n'est pas liée est dite **libre**

Liée: je peux faire un vecteur de la famille en faisant une combinaison linéaire d'autres vecteurs de la famille.

Exemple:  

Si $u_1$ et $u_2$ sont deux vecteurs de $E$, $(u_1, u_2) \text{ est une famille liée } \iff u_1 \text{ et } u_2 \text{ sont colinéaires }$

**Caractérisation** 

Soit $(u_i)_{i \in I}$ une famille de vecteurs de $E$.  

On dit que la famille $(u_i)_{i \in I}$ **est libre**  

si et seulement si:  

$\forall J \subset I \text{ fini }, \forall (\alpha_i)_{i \in J} $ famille d'éléments de $K$  

${\underset{i \in I}{\sum}} \alpha_i u_i = 0_E \Rightarrow \forall i \in J, \alpha_i = 0$

Preuve:  

On raisonne par contraposée.  

$\Rightarrow$  
On suppose: $\exists J, \exists (\alpha_i)_{i \in J}$ avec ${\underset{i \in J}{\sum}} \alpha_i u_i = 0_E$ et $\exists k \in J, \alpha_k \ne 0$  
On écrit $\alpha_k u_k = - {\underset{i \in J, i \ne k}{\sum}}\alpha_i u_i$ et on obtient $u_k = - \frac{1}{\alpha_k}{\underset{i \in J, i \ne k}{\sum}}\alpha_i u_i$  
Donc, $u_k$ s'écrit comme une combinaison linéaire des $(u_i)_{i \in J, i \ne k}$.  
On en déduit que la famille des $(u_i)_{i\in J}$ est liée.  

$\Leftarrow$  
On suppose $(u_i)_{i \in I}$ est une famille liée: $\exists k \in I, \exists K \subset I \text{ fini } \exists (\alpha_i)_{i \in K}: u_k = {\underset{i \in K}{\sum}}\alpha_i u_i$  
On pose: $J = K \cup \{k\}$ pour "passer" $k$ de l'autre côté :)  
On obtient: ${\underset{i \in J}{\sum}}\alpha_i u_i = 0_E$ avec $\alpha_k = 1 \ne 0$

Takeaway:  

**Si un vecteur s'écrit comme combinaison linéaire d'une famille libre, il s'écrit d'une seule façon.**  

On aurait pu partir de la définition suivante:  

Une famille de vecteurs $(x_1, x_2 \ldots x_n)$ est libre quand $\forall (\lambda_1, \lambda_2 \ldots \lambda_n) \in K^n, \sum \lambda_i x_i = 0_E \Rightarrow \lambda_1 = \lambda_2 = \ldots = \lambda_n = 0$

Application:  

Soient les trois vecteurs de $\mathbb{R}^3 : u = (2,-1,3), v=(-1,4,-3), w=(1,0,1)$

- Si une famille est génératrice, toute sur-famille est liée,
- Si une famille est libre, toute sous-famille est libre,
- Si une famille est libre, toute sous-famille n'est pas génératrice.

Pour la dernière propriété, si la famille est libre, aucun de ses vecteurs n'est combinaison linéaire des autres vecteurs de la famille.

## Base et coordonnées

Supposons que $B = (e_1, e_2, \ldots , e_n)$ soit une base de $E$  

Soit $u = \underset{i \in \{1 \ldots n\}}{\sum}x_i e_i$ on peut associer à $u$ ses coordonnées dans la base $B: (x_1, x_2, \ldots , x_n)$

Exemple dans la base orthonormée de $\mathbb{R}^2$ notée $(\vec{i}, \vec{j})$  

$\vec{i}\text{  et  } \vec{j}$ ont pour coordonnées respectives $(1,0)$ et $(0,1)$  

Si $\vec{u} = 1 \times \vec{i} + 2 \times \vec{j}$ alors les coordonnées de $\vec{u}$ sont $(1,2)$  

$1 \times (1, 0) + 2 \times (0, 1) = (1, 0) + (0, 2) = (1, 2)$

- Base génératrice $\Rightarrow$ chaque vecteur a des coordonnées  
- Base libre $\Rightarrow$ ces coordonnées sont uniques

**Se donner une base c'est se donner un système de coordonnées permettant de repérer tous les vecteurs.**

Définitions:  

- Une famille de vecteurs est dite minimale quand c'est la plus petite famille conservant une propriété.
- Une famille de vecterus est dite maximmale quand c'est la plus grande famille conservant une propriété.

**Caractérisation d'une base**  

Soit $(u_i)_{i \in I}$ une famille de vecteurs de $E$.  

On a les équivalences suivantes:  

$(u_i)_{i \in I}$ est une base  

$\iff$

$(u_i)_{i \in I}$ est une famille génératrice minimale  

$\iff$  

$(u_i)_{i \in I}$ est une famille libre maximale  

_Preuve:_  
Rappel d'un raisonnement par contraposée $\lnot A \implies \lnot B \iff B \implies A$.  

Supposons que $(u_i)_{i \in I}$ soit une famille génératrice:  

Si $(u_i)_{i \in I}$ n'est pas minimale, c'est une sur-famille d'une famille génératrice donc elle n'est pas libre. Ainsi libre implique minimale.  
Réciproquement si $(u_i)_{i \in I}$ est libre alors il existe un vecteur qui est combinaison linéaire des autres vecteurs, donc la famille n'est pas minimale.  
Une base est une famille génératrice minimale.  

Supposons que $(u_i)_{i \in I}$ soit une famille libre:  

Si $(u_i)_{i \in I}$ n'est pas maximale, alors nous pouvons lui rajouter un vecteur et elle restera libre. C'est donc une sous-famille libre, elle n'est donc pas génératrice.  
Si $(u_i)_{i \in I}$ n'est pas génératrice, il existe au moins un vecteur qui n'est pas combinaison linéaire des $(u_i)_{i \in I}$. Si on ajoute ce vecteur à la famille des $(u_i)_{i \in I}$ elle reste libre, donc la famille initiale n'est pas maximale.  
Une base est une famille libre maximale.  


### Théorème de la base incomplète

Soit *E* un espace vectoriel de dimension finie $n$, et soit $\{v_1, v_2, \ldots, v_k\} $ une famille libre de vecteurs dans $E$.  
Alors, il existe des vecteurs $v_{k+1}, v_{k+2}, \ldots, v_n $ tels que $\{v_1, v_2, \ldots, v_k, v_{k+1}, \ldots, v_n\} $ forme une base de $E$.

_Preuve:_  
On applique l'algorithme suivant:  

1. **Existence d'une base** :
   Soit $ \{u_1, u_2, \ldots, u_n\} $ une base de $E$.

2. **Construction de la famille complète** :
   Nous allons construire une base de $ E $ en ajoutant des vecteurs de la base $ \{u_1, u_2, \ldots, u_n\} $ à la famille libre $ \{v_1, v_2, \ldots, v_k\} $.

3. **Procédure itérative** :
   - Initialement, considérons la famille $ \{v_1, v_2, \ldots, v_k\} $.
   - Ajoutons les vecteurs de la base $ \{u_1, u_2, \ldots, u_n\} $ un par un à la famille $\{v_1, v_2, \ldots, v_k\} $ jusqu'à ce que la famille résultante soit une base de $ E $.

4. **Vérification de la liberté** :
   À chaque étape, vérifions que la famille reste libre. Si à un certain moment, l'ajout d'un vecteur $ u_i $ rend la famille dépendante (liée), alors $ u_i $ peut être exprimé comme une combinaison linéaire des vecteurs déjà présents dans la famille. Dans ce cas, nous ne l'ajoutons pas.

5. **Terminaison** :
   Le processus se termine après au plus $ n $ étapes, car $ E $ est de dimension $ n $. À la fin, nous obtenons une famille de $ n $ vecteurs qui est libre et génératrice, donc une base de \( E \).

_Conclusion:_

Ainsi, nous avons montré que toute famille libre de vecteurs dans un espace vectoriel de dimension finie peut être complétée en une base de cet espace.


# Dimension

## Théorème de la dimension

Soit $E$ un espace vectoriel de dimension finie. Alors, toutes les bases de $E$ ont le même nombre d'éléments.

_Preuve_

1. **Existence d'une base** :
   Soit $\{v_1, v_2, \ldots, v_n\}$ une base de $E$.

2. **Supposons une autre base** :
   Supposons que $\{u_1, u_2, \ldots, u_m\}$ soit une autre base de $E$.

3. **Liberté et génération** :
   - La famille $\{v_1, v_2, \ldots, v_n\}$ est libre et génératrice de $E$.
   - La famille $\{u_1, u_2, \ldots, u_m\}$ est également libre et génératrice de $E$.

4. **Procédure de remplacement** :
   Nous allons remplacer les vecteurs de la base $\{v_1, v_2, \ldots, v_n\}$ par les vecteurs de la base $\{u_1, u_2, \ldots, u_m\}$ un par un, tout en maintenant une famille libre et génératrice.

5. **Remplacement itératif** :
   - Remplaçons $v_1$ par $u_1$. Si la nouvelle famille $\{u_1, v_2, \ldots, v_n\}$ est libre, alors elle est également génératrice.
   - Continuons ce processus en remplaçant $v_2$ par $u_2$, et ainsi de suite.

6. **Terminaison** :
   - Après $m$ étapes, nous aurons remplacé $m$ vecteurs de la base $\{v_1, v_2, \ldots, v_n\}$ par les vecteurs de la base $\{u_1, u_2, \ldots, u_m\}$.
   - Si $m < n$, alors il resterait des vecteurs de la base $\{v_1, v_2, \ldots, v_n\}$ qui ne sont pas remplacés. Cela impliquerait que $\{u_1, u_2, \ldots, u_m\}$ ne génère pas $E$, ce qui est une contradiction.
   - Si $m > n$, alors nous aurions plus de vecteurs dans la base $\{u_1, u_2, \ldots, u_m\}$ que dans la base $\{v_1, v_2, \ldots, v_n\}$, ce qui impliquerait que $\{u_1, u_2, \ldots, u_m\}$ n'est pas libre, ce qui est également une contradiction.

7. **Conclusion** :
   Donc, $m = n$. Toutes les bases de $E$ ont le même nombre d'éléments.

## Lemme

Soit $n \in \mathbb{N}^{*}$ et $e_1, e_2, \ldots, e_n \in E$ un $K$-espace vectoriel. Toute famille de $n+1$ vecteurs de $\text{Vect}(e_1, e_2, \ldots, e_n)$ est liée.

**Démonstration**

1. **Définition de l'espace vectoriel engendré** :  
   L'espace vectoriel engendré par $e_1, e_2, \ldots, e_n$, noté $\text{Vect}(e_1, e_2, \ldots, e_n)$, est l'ensemble de toutes les combinaisons linéaires des vecteurs $e_1, e_2, \ldots, e_n$. i.e. :
   $\text{Vect}(e_1, e_2, \ldots, e_n) = \left\{ \sum_{i=1}^{n} \lambda_i e_i \mid \lambda_i \in K \right\}$

2. **Supposons une famille de $n+1$ vecteurs** :
   Considérons une famille de $n+1$ vecteurs $v_1, v_2, \ldots, v_{n+1}$ dans $\text{Vect}(e_1, e_2, \ldots, e_n)$.

3. **Expression des vecteurs $v_i$** :
   Chaque vecteur $v_i$ peut être exprimé comme une combinaison linéaire des vecteurs $e_1, e_2, \ldots, e_n$. C'est-à-dire, pour chaque $i \in \{1, 2, \ldots, n+1\}$, il existe des scalaires $\lambda_{i,j} \in K$ tels que :
   $v_i = \sum_{j=1}^{n} \lambda_{i,j} e_j$

4. **Combinaison linéaire des vecteurs $v_i$** :
   Considérons une combinaison linéaire des vecteurs $v_i$ :
   $\sum_{i=1}^{n+1} \alpha_i v_i = \sum_{i=1}^{n+1} \alpha_i \left( \sum_{j=1}^{n} \lambda_{i,j} e_j \right) = \sum_{j=1}^{n} \left( \sum_{i=1}^{n+1} \alpha_i \lambda_{i,j} \right) e_j$

5. **Système d'équations linéaires** :
   Pour que cette combinaison linéaire soit nulle, il faut que :
   $\sum_{j=1}^{n} \left( \sum_{i=1}^{n+1} \alpha_i \lambda_{i,j} \right) e_j = 0$
   Cela implique que :
   $\sum_{i=1}^{n+1} \alpha_i \lambda_{i,j} = 0 \quad \text{pour tout } j \in \{1, 2, \ldots, n\}$

6. **Système homogène** :
   Nous avons un système homogène de $n$ équations avec $n+1$ inconnues $\alpha_i$ :
   $\begin{cases}
   \sum_{i=1}^{n+1} \alpha_i \lambda_{i,1} = 0 \\
   \sum_{i=1}^{n+1} \alpha_i \lambda_{i,2} = 0 \\
   \vdots \\
   \sum_{i=1}^{n+1} \alpha_i \lambda_{i,n} = 0
   \end{cases}$

7. **Existence de solutions non triviales** :
   Un système homogène de $n$ équations avec $n+1$ inconnues a toujours une solution non triviale (c'est-à-dire une solution où au moins un $\alpha_i$ n'est pas nul). Cela signifie qu'il existe des scalaires $\alpha_1, \alpha_2, \ldots, \alpha_{n+1} \in K$, non tous nuls, tels que :
   $\sum_{i=1}^{n+1} \alpha_i v_i = 0$

8. **Conclusion** :
   Puisque les scalaires $\alpha_i$ ne sont pas tous nuls, la famille $\{v_1, v_2, \ldots, v_{n+1}\}$ est liée.


_Preuve du théorême de la dimension:_  

Soient $B$ et $B'$ deux bases de $E$. Notons $n=card(B), et , p=card(B')$  
Comme les $p$ vecteurs de $B'$ sont des $Vect(B)$ et $B'$ est libre par le lemme précédent $p \le n$.  
Comme les $p$ vecteurs de $B$ sont des $Vect(B')$ et $B$ est libre par le lemme précédent $n \le p$.  
Finalement $n = p$

# Somme d'espaces vectoriels

Problème de la réunion:  

Soit $E$ un $K$-espace vectoriel, $F$ et $G$ sont deux sous-espaces vectoriels de $E$, $F \cup G$ n'est pas en général un sous-espace vectoriel. L'intersection l'est.

**Somme de deux sous-espaces vectoriels:**  

**On appelle somme de $F$ et $G$ et on note $F + G$ l'espace $Vect(F \cup G)$**

$Vect(F \cup G)$ est le plus petit espace vectoriel contenant $F \cup G$.  
$Vect(F \cup G)$ est la combinaison linéaire des vecteurs de $F \cup G.$ Donc c'est une combinaison linéaire des éléments de $F$ et de $G$.  
$\implies F+G = \{f+g, f \in F, g \in G\}$

Dimension d'une somme:  

$(e_1, e_2, \ldots, e_k)$ base de $F \cap G$  

Par le théorême de la base incomplète:  
$(e_1, e_2, \ldots, e_k, f_{k+1}, f_{k+2}, \ldots, f_p)$ base de $F$  
$(e_1, e_2, \ldots, e_k, g_{k+1}, g_{k+2}, \ldots, g_q)$ base de $G$  

$\implies (e_1, e_2, \ldots, e_k, f_{k+1}, f_{k+2}, \ldots, f_p, g_{k+1}, g_{k+2}, \ldots, g_q)$ est une base de $F+G$.

En effet:  
$(e_1, e_2, \ldots, e_k, f_{k+1}, f_{k+2}, \ldots, f_p, g_{k+1}, g_{k+2}, \ldots, g_q)$ est une famille libre, car les $(e_1, e_2, \ldots, e_k, f_{k+1}, f_{k+2}, \ldots, f_p)$ forment une base de $F$ et les $(g_{k+1}, g_{k+2}, \ldots, g_q)$ ne sont pas dans $F$.  
Cette famille $(e_1, e_2, \ldots, e_k, f_{k+1}, f_{k+2}, \ldots, f_p, g_{k+1}, g_{k+2}, \ldots, g_q)$ est génératrice.

Si $dimF \lt \infty \text{ et } dimG \lt \infty:$  

$dim(F+G) = dimF + dimG - dim(F \cap G)$

$F$ est un sous-espace vectoriel de $\mathbb{R}^4$ donc est de dimension $lt 4 (\le 3)$  

Les vecteurs $(2,-1,0,0), (1,-1,0,1) \text{ et } (1,0,2,1)$ appartiennent à $F$

Donc $dimF \ge 3 \implies dimF = 3$  

Concernant $G$, les 2 premiers vecteurs ne peuvent pas faire une combinaison linéaire du troisième à cause des 0.  
Ils ne sont pas eux-mêmes colinéaires, toujours à cause des zéros.  
Donc les vecteurs de $G$ sont libres $\implies dimG = 3$. ($Vect$ est par définition générateur).  

$G \ne F$ car le vecteur par exemple $(0,1,1,0) \notin F$ $(0+2-1+0 \ne 0)$  
$\implies dim(F+G) = 4$ car $dimF = dimG = 3$ et $dim(F+G) \le dimE = 4$.  

Finalement en appliquant la formule $dim(F \cup G) = dimF + dimG - dim(F \cap G)$  

On obtient:  
$dim(F \cap G) = dimF + dimG - dim(F \cup G) = 3 + 3 - 4 = 2.$

**Somme directe:**

Alors $f - f' = g - g'$.  
Mais $f - f' \in F$ et $g - g' \in G \implies f - f'$ et $g - g' \in F \cap G$.  
$\implies f=f'$ et $g=g'$

Tout élément $x$ de $F \oplus G$ se décompose $x = f + g$ avec $f \in F, g \in G$ de manière **unique**.

Montrer que $P \oplus I = E$

- Pour $I$ : soient $f_1$ et $f_2 \in I: (f_1 + f_2)(-x) = f_1(x) + f_2(x) = -f_1(x) - f_2(x) = -(f_1 + f_2)(x) \iff f_1 + f_2 \in I$

- Idem pour $P$.  

Il faut ensuite montrer que les fonctions paires et impaires sont supplémentaires.  

- Si $f \in P \cap I, \forall x \in \mathbb{R}, f(-x) = f(x) = -f(x) \implies \forall x \in \mathbb{R}, f(x) = 0 \implies f = 0_E \implies I \cap P = \{0_E\}$.

$\forall x \in \mathbb{R}, i(-x) = -i(x)$ donc $i \in I$  
$\forall x \in \mathbb{R}, p(-x) = p(x)$ donc $p \in P$  

$p + i = f$ donc $E = P + I$  

$E = P + I, P \cap I = \{0_E\}$  

Conclusion: $E = P \oplus I$

$F \cap G = \{0_E\}$  
$dimF + dimG = dimE$.

Supposons que $F \cap G = \{0_E\}$ et que $dimF + dimG = dimE$, il faut montrer que $E = F + G$.  
Soient $(f_1,\ldots,f_p)$ une base de $F$ et $(g_1,\ldots,g_q)$ une base de $G$.  
Il suffit de montrer que $(f_1,\ldots,f_p,g_1,\ldots,g_q)$ est libre (car elle contient déjà le bon nombre d'éléments $p+q=dimE$).  
Donc si elle est libre, c'est une base.  
Pour montrer qu'elle est libre, il faut montrer que toute combinaison linéaire nulle entraîne que tous les coefficients soient nuls.  

Soient $a_1,\ldots,a_p,b_1,\ldots,b_q \in K$ tels que $a_1 f_1 + \ldots + a_p f_p + b_1 g_1 + \ldots + b_q f_q = 0_E$

Comme $(f_1, \ldots, f_p)$ est une base de $F$, $(f_1,\ldots,f_p)$ est une base de $F$, $(f_1,\ldots,f_p)$ est libre donc $a_1=\ldots=a_p=0$.  
De même $b_1=\ldots=b_q=0$.

## Codimension

Si $E = F \oplus G$ et $dimG \lt + \infty$ alors on dit que $F$ est de codimension finie et on note $codimF = dimG$  

Les sous-espaces vectoriels de codimention 1 sont appelés des hyperplans.

Dans le plan les hyperplans sont les droites.  
Dans l'espace les hyperplans sont les plans.  
Dans $\mathbb{R}^4$ les hyperplans sont les espaces de dimension 3.  
Dans $\mathbb{R}^{28}$ les hyperplans sont les espaces de dimension 27.  

En dimension infinie, les hyperplans sont ceux qui admettent un supplémentaire de dimension 1.

# Applications linéaires

Une application linéaire est une application d'un espace vectoriel dans un autre espace vectoriel.  

Soient $E, F$ deux $K$ espace vectoriels. Considérons une application de $f: E \rightarrow F$  

$f: Vect(e_1, e_2) \rightarrow Vect(f(e_1), f(e_2))$

A un vecteur l'application linéaire associe un vecteur.  

**Exemples géométriques d'applications linéaires**:  
- Symétries par rapport à un plan,
- Rotations par rapport à un axe,
- Projections sur un plan

**Linéaire**  

Soit $x = 2e_1 + e_2 \implies f(x) = 2f(e_1) + f(e_2)$

### Définition

$f: E \rightarrow F$ est dite <u>linéaire</u> si on a:  

$\forall x,y \in E, f(x + y) = f(x) + f(y)$  
$\forall x \in E, \forall k \in K, f(kx) = kf(x)$

<u>Propriétés:</u>  

$f: E \rightarrow F$ est une application linéaire si et seulement si:  
$\forall x,y \in E, \forall k \in K, f(kx + y) = kf(x) + f(y)$

__Preuve:__  

Si $f: E \rightarrow F$ est une application linéaire alors:  
- $\forall e_1, \ldots, e_p \in E, \forall a_1,\ldots, a_p \in K$ on a $f(\sum a_i e_i) = \sum a_i f(e_i)$ (preuve par récurrence)  
- $\forall x \in E, f(-x) = -f(x)$ (l'image de l'opposé d'un vecteur c'est l'opposé de l'image).  
- $f(0_E) = 0_F$

_Rappels:_  

Soient $E$ et $F$ deux $K$ espaces vectoriels de dimensions finies non nulles, et $dimE=p, dimF=n$

$\forall x_1, x_2, \ldots, x_p \in K, f(\sum x_i e_i) = \sum(x_i f(e_i))$  

Pour connaître $f$, il suffit de connaître les valeurs $f(e_1), f(e_2), \ldots, f(e_n)$


*Applications coordonnées:*  

L'application linéaire associée à $B = (e_1, e_2, \ldots, e_p)$ est:  

$\phi_B: K^p \rightarrow E$  
$(x_1, x_2, \ldots, x_p) \rightarrow \sum x_i e_i$  

- $\phi_B((x_1,x_2,\ldots,x_p) + (y_1,y_2,\ldots,y_p)) = \phi_B(x_1,x_2,\ldots,x_p) + \phi_B(y_1,y_2,\ldots,y_p)$  
- $\phi(\alpha x) = \alpha \phi(x)$

## Matrices

Lorsque $n=p$, on note $M \in M_n(A)$ (au lieu de $M_{n,n}(A)$) et on dit que $M$ est une matrice **carrée**.

_Remarque: un anneau diffère d'un corps car ses éléments ne sont pas forcément inversibles. (exemple d'anneau $\N$)_
